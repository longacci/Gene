```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### GSE112680

```python
df_gene = pd.read_csv("./final_GSE112680.csv")
df_gene
```

```python
df_gene.drop(columns='Unnamed: 0',inplace=True)
df_gene
```

```python
df_gene_T = df_gene.T
df_gene_T.columns = df_gene_T.loc['gene']
df_gene_T.drop('gene',axis=0,inplace=True)
```

```python
df_gene_T.rename(columns={"0":"label"},inplace=True)
```

```python
# Filter the sample as ALS and CON 
df_gene_filter = df_gene_T[df_gene_T["label"] != 'diagnosis: MIM']
df_gene_filter
```

```python
df_gene_filter['label'] = df_gene_filter['label'].map({"diagnosis: ALS":1,"diagnosis: CON":0})
df_gene_filter
```

```python
df_gene_filter = df_gene_filter.rename_axis(None,axis=1)
df_gene_filter = df_gene_filter.reset_index().rename(columns={"index":"Sample"})
```

```python
df_gene_filter = df_gene_filter.T.reset_index()
```

```python
df_gene_filter
```

```python
df_gene_filter.columns = df_gene_filter.iloc[0,:]
df_gene_filter = df_gene_filter.iloc[1:,:]
df_gene_filter
```

```python
df_gene_filter.iloc[:,1:] = df_gene_filter.iloc[:,1:].astype('float')
```

```python
df_gene_filter
```

```python
# Check if the data contains character 'HS.' 
sample = r"^HS+\."
df_gene_filter[df_gene_filter["Sample"].str.contains(sample)]
```

```python
df_gene_filter = df_gene_filter.groupby('Sample').agg("mean").reset_index()
```

```python
df_gene_filter
```

```python
df_gene_filter.iloc[:20,0]
```

```python
form_sp = r"\d+-[A-Za-z]{3}"
df_gene_filter = df_gene_filter[~df_gene_filter["Sample"].str.contains(form_sp)]
df_gene_filter = df_gene_filter.reset_index().drop(columns='index')
df_gene_filter
```

```python
df_gene_filter = df_gene_filter.T
df_gene_filter.columns = df_gene_filter.iloc[0]
df_gene_finnal = df_gene_filter
```

```python
df_gene_finnal = df_gene_finnal.reset_index()
df_gene_finnal = df_gene_finnal.rename(columns=df_gene_finnal.iloc[0]).drop(df_gene_finnal.index[0])
```

```python
df_gene_finnal.index.name = None
df_gene_finnal = df_gene_finnal.rename_axis(None,axis=1)
df_gene_finnal
```

```python
df_gene_finnal.to_csv("./data_preprocessing_finnal_GSE112680.csv")
```

### Combine two GSE112680 and GSE112676

```python
df_GSE112680 = pd.read_csv("./data_preprocessing_finnal_GSE112680.csv")
df_GSE112680 = df_GSE112680.drop(columns=["Unnamed: 0"])
df_GSE112676 = pd.read_csv("./data_preprocessing_finnal_GSE112676.csv")
df_GSE112676 = df_GSE112676.drop(columns=["Unnamed: 0"])
```

```python
df_GSE112676
```

```python
df_GSE112680
```

```python
gene_id80 = df_GSE112680.iloc[:,0:-1].columns
print(gene_id80)
gene_id76 = df_GSE112676.iloc[:,0:-1].columns
print(gene_id76)
```

```python
gene_id_similar = list(set(gene_id76) & set(gene_id80))
len(gene_id_similar)
```

```python
# Combine data
GSE80_general = df_GSE112680[gene_id_similar]
GSE76_general = df_GSE112676[gene_id_similar]
df_combine = pd.concat([GSE76_general,GSE80_general],axis=0).drop(columns='Sample')
df_combine
```

```python
# Get label 
df_gene_label = pd.concat([df_GSE112676["label"],df_GSE112680["label"]],axis=0).reset_index().drop(columns='index')
df_gene_label
```

```python
# Get sample
df_gene_sample = pd.concat([df_GSE112676["Sample"],df_GSE112680["Sample"]],axis=0).reset_index().drop(columns='index')
df_gene_sample
```

```python
df_combine_finnal = pd.concat([df_gene_sample.reset_index()['Sample'],df_combine.reset_index().iloc[:,1:],df_gene_label.reset_index()['label']],axis=1)
df_combine_finnal
```

```python
df_combine_finnal.to_csv("./data_preprocessing_combine_finnal.csv")
```
