```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("../data_preprocessing/data_preprocessing_combine_finnal.csv")
df.drop(columns=['Unnamed: 0'],inplace=True)
df
```

```python
feature = pd.read_csv('../Feature_selection/feature_ranks.csv')
feature
```

```python
y = df['label']
```

```python
df_important = df[feature["Gene"]]
df_important
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df_important,y,test_size=0.1,random_state=11,stratify=y,shuffle=True)
print(f"Train shape:{x_train.shape}")
print(f"Test shape:{x_test.shape}")
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
x_train_sc = scale.fit_transform(x_train)
x_test_sc = scale.transform(x_test)
```

```python
list_feature = feature['Gene']
list_feature
```

```python
ranks = []
for i in range(len(list_feature)):
    arr = []
    for j in range(i+1):
        arr.append(list_feature[j])
    ranks.append(arr)
ranks
```

```python
from sklearn.metrics import confusion_matrix,roc_auc_score
def confusion_matrix_scorer(clf=None,X=None, y=None,y_pred=None,y_prob=None):
    if clf != None:
        y_pred = clf.predict(X)
        y_prob = clf.predict_proba(X)[:,1]
    auc = roc_auc_score(y,y_prob)
    tn,fp,fn,tp = confusion_matrix(y,y_pred).ravel()
    acc = (tp+tn) / (tp+fp+fn+tn)
    precision = tp/(tp+fp)
    specificity = tn/(tn+fp)
    sensitivity = tp / (tp+fn)
    f1_score = (2*tp)/(2*tp+fp+fn)
    return {'acc':acc,'precision':precision,'specificity':specificity,'sensitivity':sensitivity,'f1_score':f1_score,'auc':auc}
```

```python
NUMBER_FOLD = 4
def create_df_val(rank,rs_cv):
    df_val = {
        "ACC":[],
        "precision":[],
        "specificity":[],
        "sensitivity":[],
        "f1_score":[],
        "AUC":[],
        "params":[]
    }
    for i in range(NUMBER_FOLD):
        str_label = f"split{i}"
        df_val["ACC"].append(np.array(rs_cv[str_label+"_test_acc"]))
        df_val["precision"].append(rs_cv[str_label+"_test_precision"])
        df_val["specificity"].append(rs_cv[str_label+"_test_specificity"])
        df_val["sensitivity"].append(rs_cv[str_label+"_test_sensitivity"])
        df_val["f1_score"].append(rs_cv[str_label+"_test_f1_score"])
        df_val["AUC"].append(rs_cv[str_label+"_test_auc"])
        df_val["params"].append(rs_cv["params"])
    df_val["ACC"] = np.array(df_val["ACC"] ).flatten()
    df_val["precision"] = np.array(df_val["precision"] ).flatten()
    df_val["specificity"] = np.array(df_val["specificity"] ).flatten()
    df_val["sensitivity"] = np.array(df_val["sensitivity"] ).flatten()
    df_val["f1_score"] = np.array(df_val["f1_score"] ).flatten()
    df_val["AUC"] = np.array(df_val["AUC"] ).flatten()
    df_val["params"] = np.array(df_val["params"] ).flatten()
    df_combine_val = pd.DataFrame(df_val).dropna(axis=0).sort_values(by="ACC",ascending=False)
    if df_combine_val.shape[0]>0:
        gene_top =  df_combine_val.iloc[0,:].values
        return [rank,gene_top[0],gene_top[1],gene_top[2],gene_top[3],gene_top[4],gene_top[5],gene_top[6]]
    else:
        return []
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV,StratifiedShuffleSplit
model = KNeighborsClassifier(n_neighbors=20, weights='uniform',p=2)

n_neighbors = [int(x) for x in np.linspace(start = 5, stop = 20, num = 16)]
weights = ['uniform', 'distance']
p = [2]

params = {
    'n_neighbors': n_neighbors,
    'weights': weights,
    'p': p
}
ss = StratifiedShuffleSplit(n_splits=NUMBER_FOLD,test_size=0.3,random_state=11)
grid_search = GridSearchCV(estimator=model, param_grid=params, 
                              cv=ss, verbose=2, return_train_score=True,refit=False,scoring=confusion_matrix_scorer)
ls_result = []
for i,rank in enumerate(ranks):
    grid_model = grid_search.fit(x_train_sc[:,:i+1],y_train)
    rs_cv = grid_model.cv_results_
    ls_result.append(create_df_val(rank,rs_cv))
```

```python
df_result_val = pd.DataFrame(ls_result)
df_result_val.columns = ["Gene","ACC","Precision","Sp","Sn","F1_score","AUC","Params"]
df_result_val.iloc[:,1:7] = round(df_result_val.iloc[:,1:7] * 100,2)
df_result_val
```

```python
df_result_val.to_csv("../Results/KNN_val.csv")
```

```python
params_op = df_result_val[df_result_val['ACC'] == max(df_result_val['ACC'])]["Params"].values[0]
gene_op = df_result_val[df_result_val['ACC'] == max(df_result_val['ACC'])]["Gene"].values
params_op
```

```python
best_model = KNeighborsClassifier(n_neighbors= params_op['n_neighbors'], weights= params_op['weights'], p= params_op['p']) 
i = len(gene_op[0])
best_model.fit(x_train_sc[:,:i],y_train)
y_pred = best_model.predict((x_test_sc[:,:i]))
y_prob = best_model.predict_proba((x_test_sc[:,:i]))[:,1]
acc = accuracy_score(y_test,y_pred)
evaluate = confusion_matrix_scorer(None,None,y_test,y_pred,y_prob)
```

```python
result_eva = [[gene_op[0],'%.2f'%(evaluate["acc"]*100),'%.2f'%(evaluate["precision"]*100),'%.2f'%(evaluate["specificity"]*100),'%.2f'%(evaluate["sensitivity"]*100),'%.2f'%(evaluate["f1_score"]*100),'%.2f'%(evaluate["auc"]*100),params_op]]
result = pd.DataFrame(result_eva)
result.columns = ['Gene','ACC','Precision','Sp','Sn','f1_Score','AUC','Parameter']
result
```
