```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from numpy import hstack
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("../data_preprocessing/data_preprocessing_combine_finnal.csv")
df.drop(columns=['Unnamed: 0'],inplace=True)
df.head()
```

```python
feature = pd.read_csv('../Feature_selection/feature_ranks.csv')
feature.head()
```

```python
y = df['label']
```

```python
df_important = df[feature["Gene"]]
df_important
```

```python
from sklearn.model_selection import train_test_split
x_train_full,x_test,y_train_full,y_test = train_test_split(df_important,y,test_size=0.5,random_state=11)
x_train,x_val,y_train,y_val = train_test_split(x_train_full,y_train_full,test_size=0.33,random_state=11)
print(f"Train shape:{x_train.shape}")
print(f"Val shape:{x_val.shape}")
print(f"Test shape:{x_test.shape}")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB

models = list()
models.append(('lr', LogisticRegression()))
models.append(('knn', KNeighborsClassifier()))
models.append(('cart', DecisionTreeClassifier()))
models.append(('svm', SVC()))
models.append(('bayes', GaussianNB()))
models
```

```python
def fit_ensemble(models,x_train,x_val,y_train,y_val):
    X = list()
    for name,model in models:
        model.fit(x_train,y_train)
        yhat = model.predict(x_val)
        yhat = yhat.reshape(len(yhat),1)
        X.append(yhat)
    X = hstack(X)
    blender = LogisticRegression()
    return blender.fit(X,y_val)
```

```python
def predict_ensemble(models,blender,x_test):
    X = list()
    for name,model in models:
        yhat = model.predict(x_test)
        yhat = yhat.reshape(len(yhat),1)
        X.append(yhat)
    X = hstack(X)
    return blender.predict(X)
```

```python
blender = fit_ensemble(models,x_train,x_val,y_train,y_val)
yhat = predict_ensemble(models,blender,x_test)
score = accuracy_score(y_test,yhat)
print(score)
```

```python
from sklearn.metrics import confusion_matrix,roc_auc_score
def cfm(y_test,y_pred):
    tn,fp,fn,tp = confusion_matrix(y_test,y_pred).ravel()
    acc = accuracy_score(y_test,y_pred)
    recall = tp/(tp+fn)
    precision = tp/(tp+fp)
    specificity = tn/(tn+fp)
    sensitivity = tp / (tp+fn)
    f1_score = round((2*recall*precision)/(recall+precision),3)
    return np.array([acc,recall,precision,specificity,sensitivity,f1_score])
```

```python
evaluate = cfm(y_test,yhat)
print(['%.2f'%(evaluate[0]*100)+'%','%.2f'%(evaluate[1]*100)+'%','%.2f'%(evaluate[2]*100)+'%','%.2f'%(evaluate[3]*100)+'%','%.2f'%(evaluate[4]*100)+'%'])
```
