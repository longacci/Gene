```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("../data_preprocessing/data_preprocessing_combine_finnal.csv")
df.drop(columns=['Unnamed: 0'],inplace=True)
df.head()
```

```python
feature = pd.read_csv('../Feature_selection/feature_ranks_mmpc.csv')
feature.head()
```

```python
y = df['label']
```

```python
df_important = df[feature["Gene"]]
df_important
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df_important,y,test_size=0.3,random_state=20)
print(f"Train shape:{x_train.shape}")
print(f"Test shape:{x_test.shape}")
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
x_train_sc = scale.fit_transform(x_train)
x_test_sc = scale.transform(x_test)
```

```python
list_feature = feature['Gene']
list_feature
```

```python
ranks = []
for i in range(len(list_feature)):
    arr = []
    for j in range(i+1):
        arr.append(list_feature[j])
    ranks.append(arr)
ranks
```

```python
from sklearn.metrics import confusion_matrix,roc_auc_score
def cfm(y_test,y_pred,y_prob):
    auc = roc_auc_score(y_test,y_prob)
    tn,fp,fn,tp = confusion_matrix(y_test,y_pred).ravel()
    acc = accuracy_score(y_test,y_pred)
    recall = tp/(tp+fn)
    precision = tp/(tp+fp)
    specificity = tn/(tn+fp)
    sensitivity = tp / (tp+fn)
    f1_score = round((2*recall*precision)/(recall+precision),3)
    mean_result = (acc+recall+precision+specificity+sensitivity+f1_score+auc)/7
    return np.array([acc,recall,precision,specificity,sensitivity,f1_score,auc,mean_result])
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

model = LogisticRegression(random_state=11)
params = {
    'penalty': ['l1', 'l2'],
    'C': [0.001, 0.01, 0.1, 1, 10,100],
    'solver': ['lbfgs', 'liblinear', 'newton-cg', 'sag', 'saga'],
    'max_iter': [100, 200, 300]
}
grid_search = GridSearchCV(estimator=model, param_grid=params, 
                              cv=5, verbose=2, return_train_score=True)
acc_max_rank = None
acc_max = -np.inf
result = []
mean_result = []
best_paramet = None
ls_acc = []
for i,rank in enumerate(ranks):
    grid_search.fit(x_train_sc[:,:i+1],y_train)
    best_parameters = grid_search.best_params_
    best_model = grid_search.best_estimator_
    best_model.fit(x_train_sc[:,:i+1],y_train)
    y_pred = best_model.predict((x_test_sc[:,:i+1]))
    y_prob = best_model.predict_proba((x_test_sc[:,:i+1]))[:,1]
    acc = accuracy_score(y_test,y_pred)
    evaluate = cfm(y_test,y_pred,y_prob)
    ls_acc.append(evaluate[0]*100)
    mean_result.append(round(evaluate[7]*100,2))
    result.append([rank,'%.2f'%(evaluate[0]*100)+'%','%.2f'%(evaluate[1]*100)+'%','%.2f'%(evaluate[2]*100)+'%','%.2f'%(evaluate[3]*100)+'%','%.2f'%(evaluate[4]*100)+'%','%.2f'%(evaluate[5]*100)+'%','%.2f'%(evaluate[6]*100)+'%',best_parameters])
```

```python
df_result = pd.DataFrame(result)
df_result.columns = ['Gene','ACC','Recall','Precision','Sp','Sn','f1_Score','AUC','Parameter']
df_result
```

```python
df_result.to_csv("../Results/LR.csv")
```
