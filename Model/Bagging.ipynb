```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("../data_preprocessing/data_preprocessing_combine_finnal.csv")
df.drop(columns='Unnamed: 0',inplace=True)
df.head()
```

```python
feature = pd.read_csv('../Feature_selection/feature_ranks_mmpc.csv')
feature.head()
```

```python
y = df['label']
```

```python
df_important = df[feature["Gene"]]
df_important
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df_important,y,test_size=0.3,random_state=20)
print(f"Train shape:{x_train.shape}")
print(f"Test shape:{x_test.shape}")
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
x_train_sc = scale.fit_transform(x_train)
x_test_sc = scale.transform(x_test)
```

```python
list_feature = feature['Gene']
list_feature
```

```python
ranks = []
for i in range(len(list_feature)):
    arr = []
    for j in range(i+1):
        arr.append(list_feature[j])
    ranks.append(arr)
ranks
```

```python
from sklearn.decomposition import PCA
def apply_pca(x,dimension=5):
    x = pd.DataFrame(x)
    pca = PCA(n_components=dimension,random_state=11)
    pca.fit(x)
    transformed_data = pca.transform(x)
    return transformed_data
```

```python
from sklearn.metrics import confusion_matrix,roc_auc_score
def cfm(y_test,y_pred,y_prob):
    auc = roc_auc_score(y_test,y_prob)
    tn,fp,fn,tp = confusion_matrix(y_test,y_pred).ravel()
    acc = accuracy_score(y_test,y_pred)
    recall = tp/(tp+fn)
    precision = tp/(tp+fp)
    specificity = tn/(tn+fp)
    sensitivity = tp / (tp+fn)
    f1_score = round((2*recall*precision)/(recall+precision),3)
    return np.array([acc,recall,precision,specificity,sensitivity,f1_score,auc])
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier 
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV

model = BaggingClassifier(estimator=SVC(),n_estimators=11, random_state=0)
knn_clf = KNeighborsClassifier(n_neighbors=10)
lr_clf = LogisticRegression(random_state=11)
svc_clf = SVC(C=1, gamma='auto', kernel='rbf', probability=True)
params = {
    'n_estimators':[int(x) for x in np.linspace(start = 100, stop = 1500, num = 15)],
    'max_samples':[0.6, 0.7, 0.8, 0.85],
    'max_features':[0.5, 0.6, 0.7, 0.8],
    'estimator':[knn_clf, lr_clf, svc_clf]
}
grid_search = GridSearchCV(estimator=model, param_grid=params, 
                              cv=5, verbose=2, return_train_score=True)
acc_max_rank = None
acc_max = -np.inf
result = []
best_paramet = None
for i,rank in enumerate(ranks):
    # x_pca = apply_pca(x_train_sc[:,:i+1],len(rank))
    grid_search.fit(x_train_sc[:,:i+1],y_train)
    best_parameters = grid_search.best_params_
    best_model = grid_search.best_estimator_
    best_model.fit(x_train_sc[:,:i+1],y_train)
    y_pred = best_model.predict((x_test_sc[:,:i+1]))
    y_prob = best_model.predict_proba((x_test_sc[:,:i+1]))[:,1]
    acc = accuracy_score(y_test,y_pred)
    evaluate = cfm(y_test,y_pred,y_prob)
    result.append([rank,'%.2f'%(evaluate[0]*100)+'%','%.2f'%(evaluate[1]*100)+'%','%.2f'%(evaluate[2]*100)+'%','%.2f'%(evaluate[3]*100)+'%','%.2f'%(evaluate[4]*100)+'%','%.2f'%(evaluate[5]*100)+'%','%.2f'%(evaluate[6]*100)+'%',best_parameters])
```

```python
df_result = pd.DataFrame(result)
df_result.columns = ['Gene','ACC','Recall','Precision','Sp','Sn','f1_Score','AUC','Parameter']
df_result
```

```python
df_result.to_csv("../Results/Bagging.csv")
```
